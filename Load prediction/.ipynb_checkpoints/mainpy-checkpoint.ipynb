```python
import numpy as np
import pandas as pd
import random as rnd
```

```python
data = pd.read_csv("C:\\Users\\elad\\Desktop\\Kaggle\\Load prediction\\train.csv")
data = np.array(data)
np.random.shuffle(data)
```

```python
# Data completion
dataSize = data.shape[0]

for i in range(dataSize):
    if data[i,12] == 'N':
        data[i,12] = 0
    elif data[i,12] == 'Y':
        data[i,12] = 1
    else:
        print('WARNING IN DATA\'s CLASS')


# Co Applicant Income
counter = 0
for x in data[:,7]:
    if x > 0:
        counter += 1

medianCoApplicantIncome = int(np.sum(data[:,7]) / counter) + 0.0

for i in range(len(data[:,7])):
    if data[i,7] == 0:
        data[i,7] = medianCoApplicantIncome

        
# Loan Amount
counter = 0
sum = 0
for x in data[:,8]:
    if not np.isnan(x):
        counter += 1
        sum += x

medianLoanAmount = int(sum / counter) + 0.0
for i in range(len(data[:,8])):
    if np.isnan(data[i,8]):
        data[i,8] = medianLoanAmount
        
```

```python
# Cleaning Data Calculations
maxApplicantIncome = 20000
minApplicantIncome = np.min(data[:,6])
maxCoApplicantIncome = 8000
minCoApplicantIncome = np.min(data[:,7])
maxLoanAmount = 450
minLoanAmount = np.min(data[:,8])

def clean_sample(sample):
    #Gender
    if sample[1] == 'Male':
        sample[1] = 0.0
    elif sample[1] == 'Female':
        sample[1] = 1.0
    else:
        sample[1] = rnd.randint(0,1)
    
    #Married
    if sample[2] == 'No':
        sample[2] = 0.0
    elif sample[2] == 'Yes':
        sample[2] = 1.0
    else:
        sample[2] = rnd.randint(0,1) + 0.0
      
#     #Dependents
#     if sample[3] == '0':
#         sample[3] = 0.0
#     elif sample[3] == '1':
#         sample[3] = 1.0
#     elif sample[3] == '2':
#         sample[3] = 2.0
#     elif sample[3] == '3+':
#         sample[3] = 3.0
#     else:
#         sample[3] = 0.6
        
    sample[3] = 0
    #Graduated
    if sample[4] == 'Not Graduate':
        sample[4] = 0.0
    elif sample[4] == 'Graduate':
        sample[4] = 1.0
    else:
        sample[4] = rnd.randint(0,1) + 0.0
        
    #Self-Employed
    if sample[5] == 'No':
        sample[5] = 0.0
    elif sample[5] == 'Yes':
        sample[5] = 1.0
    else:
        sample[5] = rnd.randint(0,1) + 0.0
        
    #Applicant Income
    sample[6] = (sample[6] - minApplicantIncome) / (maxApplicantIncome - minApplicantIncome)
    
    #Co Applicant Income
    sample[7] = (sample[7] - minCoApplicantIncome) / (maxCoApplicantIncome - minCoApplicantIncome)
    
    # Loan Amount
    sample[8] = (sample[8] - minLoanAmount) / (maxLoanAmount - minLoanAmount)
    
    # Loan Amount Term
    if sample[9] <= 120:
        sample[9] = 0.0
    elif sample[9] == 180 or sample[9] == 240 or sample[9] == 300:
        sample[9] = 1.0
    elif sample[9] == 360 or sample[9] == 480:
        sample[9] = 2.0
    else:
        sample[9] = 2.0
   
    # Credit History
    if len(sample) == 13:
        if np.isnan(sample[10]):
            sample[10] = sample[12] + 0.0
            
    # Property Area
    if sample[11] == 'Semiurban':
        sample[11] = 0.0
    elif sample[11] == 'Urban':
        sample[11] = 1.0
    elif sample[11] == 'Rural':
        sample[11] = 2.0
    else:
        print("ERROR IN PROPERY HISTORY")
    return sample
```

```python
for x in data:
    clean_sample(x)
```

```python
print(data.shape)
data = np.delete(data,3,1)
print(data.shape)
```

```python
n = 550
X_train = data[:n,1:12]
Y_train = np.array(data[:n,12], dtype='int')
X_test = data[n:,1:12]
Y_test = np.array(data[n:,12], dtype='int')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Random Forest
random_forest = RandomForestClassifier(n_estimators=100, max_depth=10,criterion='gini',max_features=11)
random_forest.fit(X_train, Y_train)
acc_random_forest = random_forest.score(X_train, Y_train)
print("Random Forest Train accuracy :  " + str(acc_random_forest))
acc_random_forest = random_forest.score(X_test, Y_test)
print("Random Forest Test accuracy :  " + str(acc_random_forest))
random_forest.feature_importances_
```
